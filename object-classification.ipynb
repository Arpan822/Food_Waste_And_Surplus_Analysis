```python
import os
import shutil
import numpy as np

os.getcwd()
```

```python
# Define train, test, and validation directories

train_dir = 'C:\\Users\\jaymo\\Capstone\\object classification\\data\\train'
test_dir = 'C:\\Users\\jaymo\\Capstone\\object classification\\data\\test'
valid_dir = 'C:\\Users\\jaymo\\Capstone\\object classification\\data\\val'
```

```python
# Define categories

categories = ['apple', 'banana', 'mango', 'orange', 'pepper', 'potato', 'carrot', 'cucumber']
```

```python
# Function to extract fruit/veggie name from file name

def extract_category(file_name):
    # Split file name by '-'
    category_name = file_name.split('-')[0].lower()

    if category_name.startswith('f'):
        category_name = category_name[5:]
        return category_name

    elif category_name.startswith('r'):
        category_name = category_name[6:]
        return category_name

    else:
        return category_name
```

```python
def organize_images(root_dir):
    for category in categories:
        os.makedirs(os.path.join(root_dir, f'{category}_fresh'), exist_ok=True)
        os.makedirs(os.path.join(root_dir, f'{category}_rotten'), exist_ok=True)

    for file_name in os.listdir(root_dir):
        if file_name.endswith('.jpg') or file_name.endswith('.png'):  # Assuming images are jpg
            category = extract_category(file_name)
            if category in categories:
                if "fresh" in file_name:
                    shutil.move(os.path.join(root_dir, file_name), os.path.join(root_dir, f'{category}_fresh', file_name))
                elif "rotten" in file_name:
                    shutil.move(os.path.join(root_dir, file_name), os.path.join(root_dir, f'{category}_rotten', file_name))
                else:
                    print(f"Unknown freshness label for file: {file_name}")
            else:
                print(f"Unknown category for file: {file_name}")
        else:
            print(f"Images are already organized!!!")
```

```python
# Organize images

organize_images(train_dir)

organize_images(test_dir)

organize_images(valid_dir)

print("Organizing images complete!")
```
